**CODE FOR SYNTETHIC DATA GENERATION**

**INSTRUCTIONS**

- Generate 15 sentences for each themes
- Change the language to any of your choice
- The themes of generation should in: sport, education and technology
- Use any other models BUT NOT 'gpt-3.5-turbo'

```python
client = OpenAI(api_key=OPENAI_API_KEY)

SENTENCES_PER_THEME = 15
LANGUAGE = "Yoruba"

THEMES = [
    {
        "name": "sport",
        "directions": (
            "soccer clubs, football players, "
            "fans banter, bet naija, worldcup games, group stage, "
            "championship, injuries, "
            # YOU CAN ADD AS MANY DIRECTIONS AS POSSIBLE, JUST FOLLOW THE LAYOUT
            # I HAVE LEFT THE OTHER TWO AS WE SAW IN CLASS FOR EASY UNDERSTANDING
        ),
    },
    {
        "name": "education",
        "directions": (
            "matriculation, JAMB scores, convocation, "
            "classroom games, group project, "
            "examinations, assignments, "
            "handouts, study guide"
        ),
    },
    {
        "name": "technology",
        "directions": (
            "computational linguistics, speech recognition, user interface, "
            "artificial intelligence, robots, "
            "algorithms, electric vehicles, "
            "quantum computing, neural networks"
        ),
    },
    # add more themes here, following the same {"name": ..., "directions": ...} shape
]

results = []

for theme in THEMES:
    prompt = f"""
Generate {SENTENCES_PER_THEME} natural, everyday {LANGUAGE} sentences on the theme of "{theme['name']}".

The sentences should reflect realistic, contemporary usage covering topics such as:
{theme['directions']}

Guidelines:
- Vary sentence length and structure (short and long, statements and questions).
- Use natural code-switching where it would realistically occur in spoken {LANGUAGE}, if applicable.
- Avoid repeating the same phrasing or sentence opener across sentences.
- Do not number the sentences or add explanations.

Return the sentences as a JSON list of strings only, nothing else.
"""

    response = client.responses.create(
        model="gpt-5", # USE ANY OTHER AVAILABLE MODEL
        input=prompt
    )

    output = response.output_text

    try:
        sentences = json.loads(output)
    except json.JSONDecodeError:
        # fallback if the model doesn't return clean JSON: split on lines
        sentences = [s.strip("-•* \n") for s in output.split("\n") if s.strip()]

    for sentence in sentences:
        results.append({"theme": theme["name"], "sentence": sentence})

df = pd.DataFrame(results)
df.to_csv("generated_sentences.csv", index=False)

# If running in Google Colab, uncomment:
from google.colab import files
files.download("generated_sentences.csv")
```

```python
import pandas as pd
import re

# Load the generated sentences from the CSV file
try:
    df_sentences = pd.read_csv('generated_sentences.csv')
except FileNotFoundError:
    print("Error: 'generated_sentences.csv' not found. Please ensure the generation cell was run successfully.")
    exit()

# Combine all sentences into a single string for analysis
all_sentences_text = ' '.join(df_sentences['sentence'].astype(str).tolist())

# Tokenize words: convert to lowercase and remove punctuation
words = re.findall(r'\b\w+\b', all_sentences_text.lower())

# Calculate total number of words
total_words = len(words)

# Calculate number of unique words
unique_words = len(set(words))

print(f"Total number of words: {total_words}")
print(f"Total number of unique words: {unique_words}")
```
